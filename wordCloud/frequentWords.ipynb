# Calculate the frequent word
## This file checks which words are the most repeating in a text file

```python
import re
import pandas as pd 
import nltk
import operator
import matplotlib.pyplot as plt
from collections import Counter
from pathlib import Path
from stop_words import get_stop_words
from nltk.corpus import stopwords
from wordcloud import WordCloud 
from PIL import Image
```

```python
txt = Path('data.txt').read_text()
listWords= txt.replace('\n', ' ').replace('.',' ').replace(',',' ').replace('`' , ' ').replace('?' , ' ').replace('!' , ' ').replace('-' , ' ').split()
```

```python
nltk.download('stopwords')
stop_words = list(get_stop_words('en'))     
nltk_words = list(stopwords.words('english'))
stop_words.extend(nltk_words)
listCleanedWords = [words for words in listWords if not words in stop_words]
```

```python
wordsDict = dict(Counter(listCleanedWords))
```

```python
pd.Series(wordsDict).sort_values(ascending=False)[:10].plot.bar()
```

```python
wc = WordCloud(background_color="white",width=5500,height=2500, max_words=20,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(wordsDict)
plt.imshow(wc)
```
